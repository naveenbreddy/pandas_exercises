### Step 4. How many products cost more than $10.00?

```python
df.head()
```

```python
prices = [float(value[1:-1]) for value in df.item_price]
```

```python
df['item_price'] = prices
```

```python
cleaned_df = df[df['quantity'] == 1]
cleaned_df[cleaned_df['item_price']>10].item_name.nunique()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
cleaned_df1 = cleaned_df.drop_duplicates(['item_name','quantity'])
cleaned_df1[['item_price','item_name']]
```

### Step 6. Sort by the name of the item

```python
cleaned_df1.sort_values('item_name',ascending = False)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
#df[df['item_price']==44.25]

df.sort_values('item_price', ascending = False).head(1)
```

```python
df[df['item_name']=='Veggie Salad Bowl']['quantity'].sum()
```

### Step 9. How many times did someone order more than one Canned Soda?
